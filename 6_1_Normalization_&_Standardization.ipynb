**NORMALIZATION BENFITS**



*   No feature dominance due to large values

*   Exploding Gradient Problem solve

*   Faster Convergence
*   Improve Accuracy







**STANDARDIZATION BENEFITS**


*   No feature dominance due to units

*   Less Sensitive to Outlier

*   Faster Convergence



NOTE : OUPUT of standardscaler or minmax scaler is np.array

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

**DATA FETCH**

```python
from sklearn.datasets import load_iris
f=load_iris()
```

```python
df=pd.DataFrame(f.data)
```

```python
df.head()
```

```python
df.columns=f.feature_names
```

```python
df.head()
```

```python
df['target']=f.target
```

```python
df.columns
```

```python
df.head()
```

 **SPLIT DATA INTO INPUT AND OUTPUT**

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x.head()
```

```python
y.head()
```

**DATA VISUALISATION** -> To check Impact on Data After Normalization and Standardization

```python
plt.figure(figsize=(9,15))
sns.scatterplot(df,x='sepal length (cm)',y='sepal width (cm)')
plt.show()
```

```python
plt.figure(figsize=(9,15))
sns.pairplot(df,hue='target')
plt.show()
```

```python
plt.figure(figsize=(9,15))
sns.pairplot(df)
plt.show()
```

```python
plt.figure(figsize=(9,15))
sns.histplot(df)
plt.show()
```

```python
plt.figure(figsize=(9,15))
sns.countplot(df)
plt.show()
```

**NORMALIZATION**

Note -> Normalization of Data is only done on features not on target

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
x_new=scaler.fit_transform(x)
```

```python
print(x_new)
```

```python
x_new=pd.DataFrame(x_new)
```

```python
x_new.head()
```

```python
x_new.columns=f.feature_names
```

```python
x_new.head()
```

# **DATA VISUALISATION AFTER NORMALIZATION**

```python
new_df=x_new.copy()
```

```python
new_df['target']=df.target
```

```python
new_df.head()
```

SCATTERPLOT -> You can plot in Scatterplot length of petal and width is shrinked between 0 and 1 that's what Normalization Do

```python

plt.figure(figsize=(9,15))
sns.scatterplot(new_df,x='sepal length (cm)',y='sepal width (cm)')
plt.show()
```

```python
plt.figure(figsize=(9,15))
sns.pairplot(new_df)
plt.show()
```

```python
plt.figure(figsize=(9,15))
sns.pairplot(new_df,hue='target')
plt.show()
```

```python
plt.figure(figsize=(9,15))
sns.histplot(new_df)
plt.show()
```

```python
plt.figure(figsize=(9,15))
sns.countplot(new_df)
plt.show()
```

# **STANDARDIZATION**

```python
x.head()
```

```python
x_stand=x.copy()
```

```python
x_stand.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sd=StandardScaler()
```

```python
df_std=sd.fit_transform(x_stand)
```

```python
print(df_std)
```

```python
df_std=pd.DataFrame(df_std)
```

```python
df_std.head()
```

```python
df_std.columns=f.feature_names
```

```python
df_std.head()
```

```python
df_std['target']=y
```

```python
df_std.head()
```

# **DATA VISUALISATION AFTER STANDARDIZATION**

```python
plt.figure(figsize=(9,15))
sns.scatterplot(df_std,x='sepal length (cm)',y='sepal width (cm)')
plt.show()
```

```python
plt.figure(figsize=(9,15))
sns.pairplot(df_std)
plt.show()
```

```python
plt.figure(figsize=(9,15))
sns.pairplot(df_std,hue='target')
plt.show()
```

```python
plt.figure(figsize=(9,15))
sns.countplot(df_std)
plt.show()
```
