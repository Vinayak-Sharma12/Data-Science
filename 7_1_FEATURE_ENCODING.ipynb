# **ORDINAL ENCODING**


```python
x=df[['review','education']]
```

```python
x.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
o_encoder=OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
```

```python
x_after_encode=o_encoder.fit_transform(x)
```

```python
print(x_after_encode)
```
